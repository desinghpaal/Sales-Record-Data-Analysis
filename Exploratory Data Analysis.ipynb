```python
#sns.relplot() helps in visualizing statistical relationships using two common approaches: scatter plots and line plots.
```

```python
sns.relplot(x='Unit Cost', y='Unit Price', hue='Item Type', data=SalesRecord)
```

```python
#sns.pairplot () function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column.
```

```python
sns.pairplot(SalesRecord)
```

```python
#corr() function helps in finding the pairwise correlation of all columns in a dataframe.
```

```python
correlation = SalesRecord.corr()
```

```python
#sns.heatmap defines the graphical representation of data using colors to visualize the value of the matrix.
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)
```

```python
#sns.countplot() shows the counts of observations in each categorical bin using bars.
```

```python
sns.countplot(SalesRecord['Item Type'])
```

```python
#sns.displot Seaborn distplot showa a histogrammical data with a line on it
```

```python
sns.displot(SalesRecord['Item Type'])
```

```python
#sns.stripplot is a graphical data anlysis technique for summarizing a univariate data set. (single variable)
```

```python
sns.stripplot(SalesRecord['Unit Price'], SalesRecord['Unit Cost'])
```

```python
#sns.catplot shows frequencies (or optionally fractions or percents) of the categories of one, two or three categorical variables
```

```python
sns.catplot(x="Unit Cost", kind="box", data=SalesRecord)
```
